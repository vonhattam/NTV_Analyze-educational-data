```python
# Vérifier des valeurs manquantes dans les columns

def nan_check_columns(df):
    
    result = pd.DataFrame(index = df.columns, columns = ['nombre de nan', 'pourcentage de nan (%)'])

    
    for i in range(len(df.columns)):
        
        result.iloc[i, 0]= df.iloc[:, i].isna().sum()
        result.iloc[i, 1]=  df.iloc[:, i].isna().sum()/df.shape[0]*100
    
    return result

nan_check_columns(EdStatsData)
```

```python
# Vérifier des valeurs manquantes dans les lignes

def nan_check_row(df):
    
    result = pd.DataFrame(index = df.index, columns = ['nombre de nan', 'pourcentage de nan (%)'])

    
    for i in range(df.shape[0]):
        
        result.iloc[i, 0]= df.iloc[i, :].isna().sum()
        result.iloc[i, 1]=  df.iloc[i, :].isna().sum()/df.shape[1]*100
    
    return result
```

```python
# Après choisir manuellement des indicateurs pertinents et intéressants pour ce projet,
# créer un DataFrame contenant le nom, le code, l'unité, le groupe de ces indicateurs

topics = pd.DataFrame(columns = ['codes', 'names', 'units', 'groups'])
codes =['IT.NET.USER.P2', 'NY.GDP.PCAP.CD', 'NY.GNP.PCAP.CD', 'SP.SEC.TOTL.IN', 'SP.TER.TOTL.IN', 'SE.SEC.ENRR', 'SE.TER.ENRR']
names =['taux d\'utilisation d\'internet', 'PIB par habitans', 'revenu par habitans', 'population à âge au lycé', 'population à âge l\'université', 'taux d\'inscription au lycé', 'taux d\'inscription à l\'université']
units =['%', '$', '$', 'habitans', 'habitans', '%', '%']
groups = ['infrastructure', 'revenue de client', 'revenue de client', 'nombre de client potentiel', 'nombre de client potentiel', 'nombre de client potentiel', 'nombre de client potentiel']
topics.codes = codes
topics.names = names
topics.units = units
topics.groups = groups
topics
```

```python
# Ajouter la colonne'Region' à la DataFrame EdStatsData, puis nommer data

country_region = EdStatsCountry[['Country Code', 'Region']]
data = pd.merge(EdStatsData, country_region, on= 'Country Code')
data = data[data['Region'].notna()]

# Sélectionner les données correspondant aux indicateurs pertinents

data = data[data['Indicator Code'].isin(codes)]
data.reset_index(drop = True, inplace = True)

# Supprimer les colonnes d'année possédant pleusieurs de valeur manquante (pourcentage de nan > 30 %)

for column in data.columns:
    if data['{}'.format(column)].isna().sum()/data.shape[0] > 0.3:
        del data['{}'.format(column)]

nan_check_columns(data)    
```

```python
# Supprimer les ligne des pays possédant pleusieurs de valeur manquante (pourcentage de nan > 20 %)

data.reset_index(drop = True, inplace = True)
drop_index = []
for i in data.index:
    if (data.iloc[i, :].isna().sum()/data.shape[1]) > 0.2 :
        drop_index.append(i)
data = data.drop(drop_index)

nan_check_row(data).sort_values('pourcentage de nan (%)').tail(30)
```

```python


# Définir les méthodes de normalisation

normalize_minmax = lambda x: (x-x.min())/(x.max() -x.min())

normalize_sigmoid = lambda x: 1/(1 + np.exp(-6*(x - x.median())/(x.quantile(0.75) - x.median())))


```

```python
# Définir la fonction pour calculer le score de diférente région 
# suivant les trois critères d'infrastructure, de revenue de client potentiel,
# et de nombre de client potentiel

def score_infrastructure_region(df):

    for i in range(topics.shape[0]):

        if df.iloc[i, 3] in 'infrastructure':

            select = data[data['Indicator Code'] == topics.iloc[i, 0]]
            region = select.groupby('Region').mean()        

            
            score = region.apply(normalize_minmax)
                       
    return score


        
def score_revenue_region(df):
    list_score = []
    
    for i in range(topics.shape[0]):

        if df.iloc[i, 3] =='revenue de client':
            
            select = data[data['Indicator Code'] == topics.iloc[i, 0]]
            region = select.groupby('Region').mean()        

            
            score = region.apply(normalize_sigmoid)
            
            list_score.append(score)
            
    score = (list_score[0] + list_score[1])/2
    
    return score



def score_client_region(df):
    list_score = []
    for i in range(topics.shape[0]): 
       
        if df.iloc[i, 3] =='nombre de client potentiel':
            
            select = data[data['Indicator Code'] == topics.iloc[i, 0]]
            region = select.groupby('Region').mean()        
   
            list_score.append(region)
    
    score = (list_score[0] * list_score[2]/100 + list_score[1] * list_score[3]/100)/2
    
    
    score = score.apply(normalize_sigmoid)
          
    return score
```

```python
# Définir la fonction pour calculer le score de différent pays 
# suivant les trois critères d'infrastructure, de revenue de client potentiel 
# et de nombre de client potentiel

def score_infrastructure_country(df):

    for i in range(topics.shape[0]):

        if df.iloc[i, 3] in 'infrastructure':

            select = data[data['Indicator Code'] == topics.iloc[i, 0]]
            select.set_index ('Country Name', drop = True, inplace = True)
            select.drop(select.columns[[0, 1, 2, -1]], axis = 1, inplace = True)
            
            score = select.apply(normalize_minmax)
            
    return score


def score_revenue_country(df):
    list_score =[]
    for i in range(topics.shape[0]):

        if df.iloc[i, 3] =='revenue de client':
            
            select = data[data['Indicator Code'] == topics.iloc[i, 0]]
            select.set_index ('Country Name', drop = True, inplace = True)
            select.drop(select.columns[[0, 1, 2, -1]], axis = 1, inplace = True)
            

            score = select.apply(normalize_sigmoid)
            list_score.append(score)
            
    score = (list_score[0] + list_score[1])/2     

    return score


def score_client_country(df):
    list_score = []
    for i in range(topics.shape[0]): 
       
        if df.iloc[i, 3] =='nombre de client potentiel':
            
            select = data[data['Indicator Code'] == topics.iloc[i, 0]]
            select.set_index ('Country Name', drop = True, inplace = True)
            select.drop(select.columns[[0, 1, 2, -1]], axis = 1, inplace = True)
            
            list_score.append(select)

    score = (list_score[0] * list_score[2]/100 + list_score[1] * list_score[3]/100)/2
    
    score = score.apply(normalize_sigmoid)
         
    return score
```

```python
# Définir une fonction pour calculer le score total en fonction de coefficient des trois critères 

def score_general_region(df, x, y, z):
    infrastructure = score_infrastructure_region(df)
    revenue = score_revenue_region(df)
    client = score_client_region(df)
    score = (infrastructure * x + revenue * y + client * z)/(x + y +z)
    
    return score


def score_general_country(df, x, y, z):
    infrastructure = score_infrastructure_country(df)
    revenue = score_revenue_country(df)
    client = score_client_country(df)
    score = (infrastructure * x + revenue * y + client * z)/(x + y + z)
    
    return score  
```

```python
# Définire une fonction pour tracer l'évolution de score total des régions

def lineplot_score_region(df):
    fig_df = df.T.plot()
    fig =fig_df.get_figure()
    fig.set_size_inches(10, 7)
    plt.xlabel('Year', fontsize = 24)
    plt.xticks(fontsize = 18)
    plt.ylabel('score', fontsize = 24)
    plt.yticks(fontsize = 18)
    plt.title('Evolution', fontsize = 30)
    
# tracer l'évolution de score total des régions en utilisant les coefficients des trois critères:
# infrastructure = 1
# revunue de client potentiel = 1
# nombre de client potentiel = 1

lineplot_score_region(score_general_region(topics, 1, 1, 1))
```

```python
# Définir une fonction pour montrer le score de trois critères et le Score général
# en année choisie

def score_general_country_year(df, x, y, z, year):
    infrastructure = score_infrastructure_country(df)
    infrastructure = pd.DataFrame(infrastructure['{}'.format(year)])
    infrastructure = infrastructure.rename(columns = {'{}'.format(year):'infrastructure'})
    infrastructure = infrastructure
    
    revenue = score_revenue_country(df)
    revenue = pd.DataFrame(revenue['{}'.format(year)])
    revenue = revenue.rename(columns = {'{}'.format(year):'revenue'})
    revenue = revenue
    
    client = score_client_country(df)
    client = pd.DataFrame(client['{}'.format(year)])
    client = client.rename(columns = {'{}'.format(year):'client'})
    client = client
    
    score = pd.concat([infrastructure, revenue, client], axis = 1)
    score['score'] = (score['infrastructure']* x + score['revenue']*y +score['client']*z)/(x+y+z)
    score = score.sort_values('score', ascending = False)
    
    
    return score

score_general_country_year(topics, 1, 1, 1, 2015).head(20)
```

```python
# Définir une fonction pour calculer le score moyen des pays dans une période choicies (year_N - year_M)
# en utilisant une combinaision de coefficient (x, y, z)

def score_mean_country(df, year_N, year_M, x, y, z):
    infrastructure = score_infrastructure_country(df)
    revenue = score_revenue_country(df)
    client = score_client_country(df)
    
    infrastructure_describe = infrastructure.loc[:, '{}'.format(year_N):'{}'.format(year_M)].T.describe()
    infrastructure_mean = pd.DataFrame(infrastructure_describe.iloc[1, :])
    
    revenue_describe = revenue.loc[:, '{}'.format(year_N):'{}'.format(year_M)].T.describe()
    revenue_mean = pd.DataFrame(revenue_describe.iloc[1, :])
    
    client_describe = client.loc[:, '{}'.format(year_N):'{}'.format(year_M)].T.describe()
    client_mean = pd.DataFrame(client_describe.iloc[1, :])
    
    score_general = x * infrastructure_mean + y * revenue_mean + z * client_mean
    
    score = pd.concat([infrastructure_mean, revenue_mean, client_mean, score_general], axis = 1)
    
    score.columns= ['infrastructure', 'revenue', 'client', 'score general']
    
    return score



# Définir une fonction pour montrer top pays classée par le score moyen
# dans une période choicies (year_N - year_M)
# en utilisant une combinaision de coefficient (x, y, z)

def rank_mean_country(df, year_N, year_M, x, y, z):
    
    score = score_mean_country(df, year_N, year_M, x, y, z)
    
    rank = list(score.sort_values('score general', ascending = False).head(10).index)
       
    return rank

# Définir une fonction pour tracer les scores moyens de trois critères
# dans une période choisie (year_N -year_M)
# de top pays classé par le score général calculé en utilisant la combinaison de coefficient (x, y, z)

def barplot_divided_score(df, year_N, year_M, x, y, z):
    ranking = rank_mean_country(df, year_N, year_M, x, y, z)
    score = score_mean_country(df, year_N, year_M, x, y, z)
    score = score.iloc[:, :-1]
    fig_score = score.loc[ranking, :].plot.bar(stacked = True)
    fig =fig_score.get_figure()
    fig.set_size_inches(10, 7)
    plt.ylabel('score')
    plt.xticks(fontsize = 18)
    plt.xlabel('Country', fontsize = 25)
    plt.legend(fontsize = 18)
    plt.ylabel('Score', fontsize = 25)
    plt.yticks(fontsize = 18)
```

```python
# plot les scores moyens de trois critères
# dans une période 2005-2015
# de top pays classé par le score général calculé en utilisant la combinaison de coefficient (1, 1, 1)
    
    
barplot_divided_score(topics, 2005, 2015, 1, 1, 1)
```

```python
# Définir une fonction pour choisir top pays classé par une méthode optimisée
# basé sur les scores moyens dans une période choisie (year_N, year_M)

# Méthode optimisé: 
        # étape 1: choisir des pays plus fort en deux critères d'infrastructure et de revenue
                # par score supérieur la quantile (x)
        # étape 2: classer ces pays choisis par la critère de nombre de client


def score_optimized_country(df, year_N, year_M, x):
    infrastructure = score_infrastructure_country(df)
    revenue = score_revenue_country(df)
    client = score_client_country(df)
    
    infrastructure_describe = infrastructure.loc[:, '{}'.format(year_N):'{}'.format(year_M)].T.describe()
    infrastructure_mean = pd.DataFrame(infrastructure_describe.iloc[1, :])
    
    revenue_describe = revenue.loc[:, '{}'.format(year_N):'{}'.format(year_M)].T.describe()
    revenue_mean = pd.DataFrame(revenue_describe.iloc[1, :])
    
    client_describe = client.loc[:, '{}'.format(year_N):'{}'.format(year_M)].T.describe()
    client_mean = pd.DataFrame(client_describe.iloc[1, :])
    
    
    score = pd.concat([infrastructure_mean, revenue_mean, client_mean], axis = 1)    
    score.columns= ['infrastructure', 'revenue', 'client']
    
    score = score[score.infrastructure > score.infrastructure.quantile(x)]
    score = score[score.revenue > score.revenue.quantile(x)]    
    
    return score





# Définir une fonction pour montrer top pays classé par une méthode optimisée
# basé sur les scores moyens dans une période choisie (year_N, year_M)


def rank_optimized_country(df, year_N, year_M, x):
    
    score = score_optimized_country(df, year_N, year_M, x)
    rank = list(score.sort_values('client', ascending = False).head(10).index)
    
    return rank




# Définir une fonction pour plot le score de trois critères de top pays classé par la méthode optimisée

def barplot_rank_optimized_country(df, year_N, year_M, x):
    ranking = rank_optimized_country(df, year_N, year_M, x)
    score = score_optimized_country(df, year_N, year_M, x)

    fig_score = score.loc[ranking, :].plot.bar()
    fig =fig_score.get_figure()
    fig.set_size_inches(10, 7)
    plt.ylabel('score')
    plt.xticks(fontsize = 18)
    plt.xlabel('Country', fontsize = 25)
    plt.legend(fontsize = 18)
    plt.ylabel('Score', fontsize = 25)
    plt.yticks(fontsize = 18)
```

```python
# Plot le score de trois critères de top pays dans la période 2005-2015 classé par la méthode Optimisation (0.7)
# basé sur la quantile x = 0.7

barplot_rank_optimized_country(topics, 2005, 2015, 0.7)
```

```python
# Définir une fonction pour choisir top pays classé par une méthode optimisée
# basé sur le score de trois critères dans une année choisie

# Méthode optimisé: 
        # étape 1: choisir des pays plus fort en deux critères d'infrastructure et de revenue
                # par score supérieur la quantile (x)
        # étape 2: classer ces pays choisis par la critère de nombre de client


def score_optimized_country_year(df, year, x):
    infrastructure = score_infrastructure_country(df)
    infrastructure = pd.DataFrame(infrastructure['{}'.format(year)])
    infrastructure = infrastructure.rename(columns = {'{}'.format(year):'infrastructure'})
    infrastructure = infrastructure
    
    revenue = score_revenue_country(df)
    revenue = pd.DataFrame(revenue['{}'.format(year)])
    revenue = revenue.rename(columns = {'{}'.format(year):'revenue'})
    revenue = revenue
    
    client = score_client_country(df)
    client = pd.DataFrame(client['{}'.format(year)])
    client = client.rename(columns = {'{}'.format(year):'client'})
    client = client
    
    score = pd.concat([infrastructure, revenue, client], axis = 1)

    score = score[score.infrastructure > score.infrastructure.quantile(x)]
    score = score[score.revenue > score.revenue.quantile(x)]
    score = score.sort_values('client', ascending = False)
    
    return score

score_optimized_country_year(topics, 2015, 0.7)
```
